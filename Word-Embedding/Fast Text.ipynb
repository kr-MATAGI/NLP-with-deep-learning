# 패스트 텍스트(FastText)

- 페이스북에서 개발한 단어를 벡터로 만드는 방법.
- Word2Vec 이후에 나온 것이기에 메커니즘 자체는 Word2Vec의 확장이라고 볼 수 있다.
- Word2Vec와 Fast Text의 가장 큰 차이점이라면 <br> Word2Vec는 단어를 쪼개질 수 없는 단위로 생각한다면, <br> Fast Text는 하나의 단어 안에도 여러 단어들이 존재하는 것으로 간주한다. <br> 즉, <b>내부 단어(subword)</b>를 고려하여 학습한다.

## 1. 내부 단어(subword) 학습

- FastText에서는 각 단어는 글자 단위 n-gram의 구성으로 취급한다. <br> n을 몇으로 결정하는지에 따라서 단어들이 얼마나 분리되는지 결정된다.

- <b>이런 방법은 Word2Vec에서는 얻을 수 없었던 강점을 가진다.</b>

## 2. 모르는 단어(Out of Vocabulary, OOV)에 대한 대응
- FastText의 인공 신경망을 학습한 후에는 데이터 셋의 모든 단어의 각 n-gram에 대해서 워드 임베딩이 된다. <br> 이렇게 되면 장점은 데이터 셋만 충분하다면 위와 같은 내부 단어를 통해 모르는 단어(OOV)에 대해서도 다른 단어와의 유사도를 계산할 수 있다.



## 3. 단어 집합 내 빈도 수가 적었던 단어(Rare Word)에 대한 대응

- Word2Vec의 경우에는 등장 빈도 수가 적은 단어(Rare Word)에 대해서는 임베딩의 정확도가 높지 않다는 단점이 있었다. <br> 참고할 수 있는 경우의 수가 적다보니 정확하게 임베딩이 되지 않는 경우이다.

- FastText의 경우, 만약 단어가 희귀 단어라도 그 단어의 n-gram이 다른 단어의 n-gram과 겹치는 경우라면, Word2Vec와 비교하면 비교적 높은 임베딩 벡터값을 얻을 수 있다.
<br><br>
- FastText가 노이즈가 많은 코퍼스에서 강점을 가진 것 또한 이와 같은 이유이다. <br> 모든 훈련 코퍼스에서 오타(Typo)나 맞춤법이 틀린 단어가 없으면 이상적이겠지만, 실제 많은 비정형 데이터에는 오타가 섞여 있다. <br> 그리고 오타가 섞인 단어는 당연히 등장 빈도수가 매우 적으므로 일종의 희귀 단어가 된다.
- 즉, Word2Vec에서는 오타가 섞인 단어는 임베딩이 제대로 되지 않지만 FastText는 이에 대해서도 일정 수준의 성능을 보인다.
- 예를 들어 단어 apple과 오타로 p를 한 번 더 입력한 appple의 경우에는 실제로 많은 개수의 동일한 n-gram을 가질 것입니다.
